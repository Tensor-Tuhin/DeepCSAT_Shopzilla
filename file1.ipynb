1. PROJECT TITLE - DEEPCSAT – ECOMMERCE CUSTOMER SATISFACTION SCORE PREDICTION.

2. PROJECT SUMMARY - 

    The project involved data cleaning, feature engineering, and encoding categorical variables using one-hot encoding. Exploratory Data Analysis (EDA) was conducted to identify patterns in customer interactions and satisfaction levels. Key insights revealed that returns and order-related queries generated the highest interaction volumes and that CSAT scores were highly imbalanced toward the highest rating.

    A neural network model was developed using TensorFlow and Keras to treat CSAT prediction as a multi-class classification problem. The model consisted of multiple dense layers with ReLU activation and a Softmax output layer to predict the five CSAT classes. Class weights were applied to address class imbalance during training.

    Model performance was evaluated using accuracy, precision, recall, F1-score, and confusion matrix analysis. The trained model was then prepared for deployment through a Streamlit application, enabling real-time CSAT prediction based on input interaction features.

    This project demonstrates an end-to-end machine learning workflow, including data preprocessing, exploratory analysis, model development, evaluation, and deployment for practical customer support analytics.

3. PROBLEM STATEMENT- 

    This project focuses on predicting Customer Satisfaction (CSAT) scores using Deep Learning Artificial Neural Networks (ANN). In the context of e-commerce, understanding customer satisfaction through their interactions and feedback is crucial for enhancing service quality, customer retention, and overall business growth. By leveraging advanced neural network models, we aim to accurately forecast CSAT scores based on a myriad of interaction-related features, providing actionable insights for service improvement.

4. LOADING AND CHECKING THE DATASET.

```python
# Importing the necessary libraries
import pandas as pd
import warnings
import os
warnings.filterwarnings('ignore')
```

```python
# To prevent truncating
pd.set_option('display.max_columns', None)

# Loading the dataset
df_unclean = pd.read_csv('../unclean_data/eCommerce_Customer_support_data.csv')

# Checking few sample rows of the dataframe
df_unclean.sample(7)
```

```python
# Checking the info of df_unclean
df_unclean.info()
```

```python
# Checking the description of df_unclean
df_unclean.describe()
```

```python
# Checking the null values of df_unclean
df_unclean.isna().sum()
```

```python
# Removing columns that are either very sparse or those that won't be useful for predicting CSAT
df_unclean = df_unclean.drop(columns=['Unique id', 'Customer Remarks', 'Order_id', 'order_date_time', 'Customer_City', 'Product_category',
                                      'Item_price', 'connected_handling_time', 'Agent_name', 'Supervisor', 'Manager'])

# Checking rows of the updated dataframe
df_unclean.sample(5)
```

```python
# Rechecking for null values
df_unclean.isnull().sum()
```

```python
# Rechecking the info
df_unclean.info()
```

```python
# Changing the dtypes of Issue_reported at, issue_responded and Survey_response_Date columns
df_unclean['Issue_reported at'] = pd.to_datetime(df_unclean['Issue_reported at'], dayfirst=True)
df_unclean['issue_responded'] = pd.to_datetime(df_unclean['issue_responded'], dayfirst=True)
df_unclean['Survey_response_Date'] = pd.to_datetime(df_unclean['Survey_response_Date'], dayfirst=True)
```

```python
# Checking the dtypes after changing
df_unclean.dtypes
```

```python
# Checking the first few rows
df_unclean.head()
```

```python
# Creating a new feature "responsetime"
df_unclean['responsetime'] = df_unclean['issue_responded'] - df_unclean['Issue_reported at']
df_unclean.head()
```

```python
# Changing the values of the 'responsetime' column to minutes
df_unclean['responsetime_in_mins'] = df_unclean['responsetime'].dt.total_seconds()/60

# Keeping only the rows that do not have any negative values in the 'responsetime_in_mins' column
df_unclean = df_unclean[df_unclean['responsetime_in_mins'] >= 0]

# Checking first few rows
df_unclean.head()
```

```python
# Ordinally encoding the 'Tenure Bucket' column
tenure = {'On Job Training': 0,
          '0-30': 1,
          '31-60': 2,
          '61-90': 3,
          '>90': 4}

df_unclean['Tenure_Bucket_enc'] = df_unclean['Tenure Bucket'].map(tenure)

# Checking the first few rows
df_unclean.head()
```

```python
# Checking the info of the updated dataframe
df_unclean.info()
```

```python
# Fixing the capitalisation issues of 'category' column
df_unclean['category'] = df_unclean['category'].str.title()

# Changing the dtype of 'responsetime_in_mins' column to int64
df_unclean['responsetime_in_mins'] = df_unclean['responsetime_in_mins'].astype(int)
```

```python
# Omitting the columns that won't be necessary
df_clean = df_unclean.drop(columns=['Issue_reported at', 'issue_responded', 'Survey_response_Date', 'Tenure Bucket', 'responsetime'])

# Checking the rows of the new dataframe
df_clean.head()
```

```python
# Making a folder to store the cleaned dataframe
os.makedirs('../clean_data', exist_ok=True)
```

```python
# Storing both the clean and the unclean dataframe
df_unclean.to_csv('../unclean_data/df_unclean.csv', index=False)
df_clean.to_csv('../clean_data/df_clean.csv', index=False)
```
