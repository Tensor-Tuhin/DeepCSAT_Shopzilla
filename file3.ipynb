In this file we will be creating dummy columns from the categorical columns and then we will train an Artificial Neural Network model to predict the Customer Satisfaction Score Finally, we will save the trained model.

First we will load the df_clean dataframe.

```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
# Loading the dataset
df = pd.read_csv('../clean_data/df_clean.csv')

# Checking few rows
df.head()
```

We will be encoding the columns of channel_name, category, Sub-category and Agent Shift by creating dummy columns.

```python
# Creating the dummy columns
df_enc = pd.get_dummies(df, columns=['channel_name', 'category', 'Sub-category', 'Agent Shift'], drop_first=False, dtype=int)

# To prevent truncating
pd.set_option('display.max_columns', None)

# Checking few rows of this updated dataframe
df_enc.head()
```

```python
# Checking the name of the columns of df_enc
df_enc.columns.to_list()
```

We will separate the data into training and testing data.

```python
# Defining the features and target variable
x = df_enc.drop('CSAT Score', axis=1)
y = df_enc['CSAT Score']
```

```python
# Importing the library
from sklearn.model_selection import train_test_split

# Train-Test split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.3, stratify=y)

# Since the target classes are 1-5, we will change them to 0-4
y_train = y_train - 1
y_test = y_test - 1

# Checking the classes
print(y_train.unique())
print(y_test.unique())

# Checking the shape of the splits
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

We will now scale the column of responsetime_in_mins using StandardScaler.

```python
# Initialising the scaler
scaler = StandardScaler()

# Scaling
x_train[['responsetime_in_mins']] = scaler.fit_transform(x_train[['responsetime_in_mins']])
x_test[['responsetime_in_mins']] = scaler.transform(x_test[['responsetime_in_mins']])
```

We will now initialise the model and train it.

```python
# Importing the libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Building the neural network
m = Sequential()

m.add(Input(shape=(79,)))                   # Input layer consisting of 79 neurons, equivalent to the number of features

m.add(Dense(256, activation='relu'))        # Hidden layer
m.add(Dropout(0.3))    # Prevents overfiting by randomly disabling 30% of neurons, so that the neurons don't depend on each other and learn independently

m.add(Dense(128, activation='relu'))         # Hidden layer
m.add(Dropout(0.3))    # Dropout layer

m.add(Dense(64, activation='relu'))         # Hidden layer

m.add(Dense(5, activation='softmax'))       # Output layer consisting of 5 neurons, equivalent to the number of classes of the target

# Softmax is used in output layer for probabilty distribution and multi-class classification
# Relu (Rectified Linear Unit) is used in the hidden layers due to it's efficiency and reduced vanishing gradient issues
```

```python
# Compiling the model
m.compile(optimizer=Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Early stopping to avoid overfitting
early_stop = EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True)

# Model checkpoint - to save the best model
checkpoint = ModelCheckpoint("best_model.keras", monitor='val_accuracy', save_best_only=True, verbose=1)
```

```python
# Computing class weights to make the model learn better for minority classes
from sklearn.utils.class_weight import compute_class_weight

classes = np.unique(y_train)
weights = compute_class_weight(classes=classes, class_weight='balanced', y=y_train)
class_weights = dict(enumerate(weights))
```

```python
# Training the model
history = m.fit(x_train, y_train,
                epochs=50, batch_size=32,
                callbacks=[early_stop, checkpoint],
                validation_split=0.2, class_weight=class_weights)
```

```python
# Evaluation
test_loss, test_accuracy = m.evaluate(x_test, y_test)
print("Test accuracy:", test_accuracy)
print("Test loss:", test_loss)
```

```python
# Predicting the results
y_pred_prob = m.predict(x_test)
y_pred = y_pred_prob.argmax(axis=1)
```

```python
# Performance metrics
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
# Saving the scaler, df_enc dataframe and the list of columns
import os
import joblib

os.makedirs('../artifacts', exist_ok=True)
joblib.dump(scaler, '../artifacts/scaler.pkl')
joblib.dump(x_train.columns.to_list(), '../artifacts/columns.pkl')
df_enc.to_csv('../clean_data/df_enc.csv')
```
